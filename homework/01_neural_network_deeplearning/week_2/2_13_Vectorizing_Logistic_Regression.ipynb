### <font color=#0099ff>重点总结：</font>

所有 m 个样本的线性输出 $Z$ 可以用矩阵表示：

$Z = w^{T}X+b$

```
Z = np.dot(w.T,X) + b
A = sigmoid(Z)
```

**逻辑回归梯度下降输出向量化**

- $dZ$对于$m$个样本，维度为$（1,m）$，表示为： 

	<center>$dZ = A - Y$</center>

- db可以表示为： 

<center>$db = \dfrac{1}{m}\sum_{i=1}^{m}dz^{(i)}$</center>

```
 db = 1/m * np.sum(dZ)
```

- dw可表示为： 

<center>$dw = \dfrac{1}{m}X\cdot dZ^{T}$</center>

```
dw = 1/m*np.dot(X,dZ.T)
```
